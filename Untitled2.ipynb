IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading dataset "weatherAUS.csv" into "a"

```python
a= pd.read_csv('weatherAUS.csv')
a.head()
```

```python
a.info()
```

MISSING DATA ANALYSIS

```python
missing = round((a.isnull().sum())/(a.isnull().count())*100,1)
missing.sort_values(ascending=False).head()
```

```python
sns.heatmap(data=a.isnull())
```

```python
sns.heatmap(data=a.isnull(), yticklabels=False, 
            cbar=False, cmap='viridis' )
```

Here, yellow represent the data whichis missing!

Now, DATA CLEANING:

Here is the missing data, we can drop these columns!

```python
a.drop(['Sunshine','Evaporation','Cloud9am','Cloud3pm','Date','Location'],axis=1,inplace=True)
```

```python
a.dropna(inplace=True)
sns.heatmap(a.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

Now, We don't have any missing data in our dataset now!

```python
a.head(5)
```

To apply Machine Learning algorithm, we need to convert categorical variable into some indicator variables

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a['WindGustDir']= le.fit_transform(a['WindGustDir'])
a['WindDir9am']= le.fit_transform(a['WindDir9am'])
a['WindDir3pm']= le.fit_transform(a['WindDir3pm'])
a['RainToday']= le.fit_transform(a['RainToday'])
a['RainTomorrow']= le.fit_transform(a['RainTomorrow'])
```

```python
a.head(3)
```

Model Building - Logistic Regression

```python
x= a.drop(['RainTomorrow'],axis=1)
y=a['RainTomorrow']
```

```python
plt.figure(figsize = (6,6))
sns.scatterplot(x='MaxTemp',y='MinTemp',hue= 'RainTomorrow', palette ='inferno', data=a)
```

```python
plt.figure(figsize = (6,6))
sns.scatterplot(x='Humidity9am',y='Temp9am',hue= 'RainTomorrow', palette ='inferno', data=a)
```

HEATMAP -- CORRELATION

```python
plt.figure(figsize = (6,6))
sns.heatmap(a.corr())
```

TRAINING

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
```

```python
logR.fit(x_train, y_train)
```

PREDICTIONS

```python
predictions = logR.predict(x_test)
```

```python
predictions
```

MESURUING MODEL PERFORMANCE

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
score = logR.score(x_test, y_test)
print(score)
```

The logistic Regression model accuracy score is 0.84
